## Tutorial 29 Basic image processing

```python
from skimage import io, color
```

```python
def import_img(path, as_gray=False):
    img = io.imread(path, as_gray = as_gray)
    plt.imshow(img)
    return img
```

## Tutorial 31 Image filtering 

Unsharpened image = original * (original - blurred)

step 1 : Original - Blurred = Original-Blurred

step 2 : Original + (n*Original-Blurred) = Unsharpened image

```python
from skimage import io, img_as_float
from skimage.filters import unsharp_mask
from skimage.filters import gaussian
```

```python
img_f = img_as_float(img_sk1)
```

```python
gaussian_img = gaussian(img_f, sigma = 2 , mode='constant', cval= 0.0)
```

```python
img_f_2 = (img_f - gaussian_img)*2
```

```python
img_f_3 = img_f + img_f_2
```

```python
plt.imshow(img_f_3, cmap= 'gray')
```
