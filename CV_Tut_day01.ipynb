## Tutorial 24 Saving images from python

```python
plt.imshow(img_sk)
```

```python
from skimage import filters
```

```python
gaussian_img = filters.gaussian(img_sk, sigma = 1)
```

```python
plt.imshow(gaussian_img)
```

```python
from skimage import img_as_ubyte
```

```python
gaussian_img_8bit = img_as_ubyte(gaussian_img)
```

```python
io.imsave('someimg1.tif', gaussian_img)
```

```python
cv2.imwrite('someimg2.tif', gaussian_img_8bit)
```
